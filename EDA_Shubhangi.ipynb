# 2. Clean the Data

  2.1 Drop missing values

```python
health.isnull().sum()
```

```python
# Dropping Duplicates
health.drop_duplicates(inplace=True)
```

```python
health.columns
```

```python
for i in health:
    print(i,health[i].unique())
```

```python
health.columns
```

```python
# Replace the column values
health['HeartDiseaseorAttack'].replace({0:'No',1:'Yes',},inplace=True)
health['HighBP'].replace({0:'No high blood pressure',1:'High blood pressure'},inplace=True)
health['HighChol'].replace({0:'No high cholesterol',1:'High cholesterol'},inplace=True)
health['CholCheck'].replace({0:'No cholesterol check',1:'Cholesterol check'},inplace=True)
health['Smoker'].replace({0:'No smoker',1:'Smoker'},inplace=True)
health['Stroke'].replace({0:'No stroke',1:'Stroke'},inplace=True)
health['Diabetes'].replace({0:'No diabetes',1:'Prediabetes',2:'Diabetes'})
health['PhysActivity'].replace({0:'No',1:'Yes'},inplace=True)
health['Fruits'].replace({0:'No',1:'Yes'},inplace=True)
health['Veggies'].replace({0:'No',1:'Yes'},inplace=True)
health['HvyAlcoholConsump'].replace({0:'No',1:'Yes'},inplace=True)
health['AnyHealthcare'].replace({0:'No',1:'Yes'},inplace=True)
health['NoDocbcCost'].replace({0:'No',1:'Yes'},inplace=True)
health['GenHlth'].replace({1:'Excellent',2:'Very good',3:'Good',4:'Fair',5:'Poor'},inplace=True)
health['DiffWalk'].replace({0:'No',1:'Yes'},inplace=True)
health['Age'].replace({1:'18-24',2:'25-29',3:'30-34',4:'35-39',5:'40-44',6:'45-49',7:'50-54',8:'55-59',9:'60-64',10:'65-69',11:'70-74',12:'75-79',13:'80 or older'},inplace=True)
health['Education'].replace({1:'Never attended school or only kindergarten', 2:'Grades 1 through 8 (Elementary)', 3:'Grades 9 through 11 (Some high school)', 4:'Grade 12 or GED (High school graduate)',5:'College 1 year to 3 years (Some college or technical school)',6:'College 4 years or more (College graduate)'})
health['Income'].replace({1:'Less than $10,000',2:'Less than $15,000 ($10,000 to less than $15,000)',3:'Less than $20,000 ($15,000 to less than $20,000)',4:'Less than $25,000 ($20,000 to less than $25,000)',5:'Less than $35,000 ($25,000 to less than $35,000)',6:'Less than $50,000 ($35,000 to less than $50,000)',7:'Less than $75,000 ($50,000 to less than $75,000)',8:'$75,000 or more'})
health['Sex'].replace({0:'Female',1:'Male'},inplace=True)
```

# **3. Univariate Analysis**

```python
# Histogram to visualise the age
health = health.sort_values(by='Age', ascending=True)

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(health['Age'], color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')

# Display the plot
plt.show()


```

```python
#Bar chart to visualise the BMI Index
# Count how many people fall into each BMI value
bmi_counts = health['BMI'].value_counts().sort_index()

# Create the bar chart
plt.figure(figsize=(10, 5))

bars = plt.bar(bmi_counts.index, bmi_counts.values, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Bar Chart of BMI Index', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Histogram to visualise number of days mental health of person was not good in past 30 days
plt.figure(figsize=(10, 5))

# Create histogram
plt.hist(health['MentHlth'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Number of Days Mental Health Was Not Good (030)', fontsize=12)
plt.ylabel('Count of People', fontsize=12)
plt.title('Histogram of Poor Mental Health Days (Past 30 Days)', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Histogram to visualise number of days physical health of person was not good in past 30 days
plt.figure(figsize=(10, 5))

# Create histogram
plt.hist(health['PhysHlth'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Number of Days Physical Health Was Not Good (030)', fontsize=12)
plt.ylabel('Count of People', fontsize=12)
plt.title('Histogram of Poor Physical Health Days (Past 30 Days)', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
obj = [
    'HeartDiseaseorAttack', 'HighBP', 'HighChol', 'PhysActivity',
    'Fruits', 'Veggies', 'NoDocbcCost', 'GenHlth',
    'DiffWalk', 'Sex', 'HvyAlcoholConsump', 'AnyHealthcare'
]

# Choose different color palettes to rotate through
palettes = ['Set2', 'pastel', 'muted', 'husl', 'coolwarm', 'tab10']

fig, axes = plt.subplots(nrows=len(obj)//2 + len(obj)%2, ncols=2, figsize=(12, 30))
fig.suptitle("Categorical Feature Value Counts", fontsize=20)
axes = axes.flatten()

for i, (ax, col) in enumerate(zip(axes, obj)):
    sns.countplot(data=health, x=col, ax=ax, palette=palettes[i % len(palettes)])
    ax.set_title(col, fontsize=12)
    ax.set_xlabel("")
    ax.set_ylabel("Count")

# Remove empty subplots if any
for ax in axes[len(obj):]:
    ax.remove()

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
obj1 = ( 'CholCheck', 'Smoker','Stroke',"AnyHealthcare")
plt.figure(figsize=(15,15))
for i,column in enumerate(obj1):
    plt.subplot(len(obj1), 2, i+1)
    plt.suptitle("Plot Value Proportion", fontsize=20, x=0.5, y=1)
    plt.pie(x=health[column].value_counts(), labels=health[column].unique(), autopct='%.0f%%')
    plt.title(f"{column}")
    plt.tight_layout()
```

**Heart Disease Attack and Age**

```python
plt.figure(figsize=(10, 6))
sns.countplot(
    x='Age',
    hue='HeartDiseaseorAttack',
    data=health,
    palette='Set2'
)

plt.title('Heart Disease/Attack by Age', fontsize=14)
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Heart Disease Attack and Sex**

```python
plt.figure(figsize=(8, 6))
sns.countplot(
    x='Sex',
    hue='HeartDiseaseorAttack',
    data=health,
    palette='Set2'
)

plt.title('Heart Disease/Attack by Sex', fontsize=14)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**HighChol vs HighBP**

```python
heat_data = health.pivot_table(index='HighChol', columns='HighBP', aggfunc='size', fill_value=0)
plt.figure(figsize=(6, 4))
sns.heatmap(heat_data, annot=True, cmap='coolwarm', fmt='d')

plt.title('High Cholesterol vs High Blood Pressure')
plt.xlabel('High Blood Pressure')
plt.ylabel('High Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Smoker', hue='HeartDiseaseorAttack', data=health)
plt.title(' somking vs HeartDiseaseorAttack')
plt.show()
```

**Factors Affecting Heart Disease or Attack**

```python
# List of factors
factors = ['HighBP', 'HighChol', 'CholCheck', 'Smoker', 'HvyAlcoholConsump']

# Plot each factor
for factor in factors:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=factor, hue='HeartDiseaseorAttack', data=health, palette='Set2')
    plt.title(f'Heart Disease vs {factor}')
    plt.ylabel('Count')
    plt.show()
```

**Activities Affecting Heart Disease or Attack**

```python
# List of factors to plot
activities = ["PhysActivity", "Fruits", "Veggies", "AnyHealthcare", "NoDocbcCost", "DiffWalk"]

# Loop through each factor and plot
for activity in activities:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=activity, hue='HeartDiseaseorAttack', data=health, palette='Set2')
    plt.title(f'Heart Disease vs {activity}')
    plt.ylabel('Count')
    plt.show()
```

# **Multivariate Analysis**

```python
sh = ["HighBP", "HighChol"]

plt.figure(figsize=(12, 6))
plt.suptitle("High BP & High Cholesterol vs Heart Disease or Attack",
             fontsize=18, fontweight='bold', y=1.02)

for i, column in enumerate(sh):
    plt.subplot(1, len(sh), i+1)
    sns.countplot(data=health, x=column, hue="HeartDiseaseorAttack", palette="Set2")
    plt.title(f"{column} vs Heart Disease/Attack", fontsize=12)
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.legend(title="HeartDiseaseorAttack", labels=["No", "Yes"])

plt.tight_layout()
plt.show()
```

# **Line Plots**
